# chap 03
## P53
Meta-labeling is a very powerful tool to have in your arsenal, for four additional reasons.
元标签是一个非常强大的工具，有四个额外的原因。
First, ML algorithms are often criticized as black boxes. You may have read in the press that many hedge funds are embracing the quantamental approach. Meta-labeling allows you to build an ML system on top of a white box
第一, ML算法经常被批评为黑盒子。你可能在报纸上读到过,许多对冲基金正在拥抱量化基金。而元标签允许你在白盒子上构建一个ML系统
Second, the effects of overfitting are limited when you apply metalabeling, because ML will not decide the side of your bet, only the size.
第二,当你应用元标签时,过度拟合的影响是有限的,因为ML不会决定你的赌注的一面,只是大小。
Third, by decoupling the side prediction from the size prediction, meta-labeling enables sophisticated strategy structures.
第三,通过将侧预测与大小预测分离,元标签使复杂的策略结构成为可能。
Fourth, achieving high accuracy on small bets and low accuracy on large bets will ruin you.
第四,在小赌注上实现高精度,在大赌注上实现低精度会毁了你。
